```python
import pandas as pd
import numpy as np
```

### 1. Cargamos Datos

```python
dataset = pd.read_csv("../data/raw/train.csv")
dataset.head()
```

### 2. Eliminamos variables no utiles

```python
dataset.drop(['PassengerId','Name','Ticket'], axis=1, inplace=True)
dataset.head()
```

### 3. Porcentaje de Nulos

```python
dataset.isnull().mean()
```

Eliminamos la columna Cabin dado que tiene 77% de faltantes

```python
dataset.drop(['Cabin'], axis=1, inplace=True)
dataset.head()
```

Imputamos la media en la columna Age

```python
media_age = dataset['Age'].mean()
dataset['Age'] = dataset['Age'].fillna(media_age).astype(int)
```

Imputamos Embarked con valor más frecuente.

```python
dataset['Embarked'].value_counts()
```

```python
dataset['Embarked'] = dataset['Embarked'].fillna('S')
```

### 4. Codificación de Variables Categóricas

```python
dataset.head()
```

Codificación de varibles Sex con One Hot Encoding

```python
dataset['Sex'] = pd.get_dummies(dataset['Sex'], drop_first=True).astype(int)
```

Codificación de variables Embarked con Frecuency Encoding

```python
codificador_embarked = dataset['Embarked'].value_counts()
dataset['Embarked'] = dataset['Embarked'].map(codificador_embarked)
```

```python
dataset.head()
```

### 4. Estandarización de Variables

```python
#pip install scikit-learn
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = dataset.drop(['Survived'], axis=1)

mm_scaler = MinMaxScaler()
scaled_features = mm_scaler.fit_transform(X)

scaled_df = pd.DataFrame(scaled_features, columns=X.columns)
```

```python
scaled_df['Survived'] = dataset['Survived']
scaled_df.head()
```

```python
scaled_df.to_csv('../data/processed/features_for_model.csv', index=False)
```
